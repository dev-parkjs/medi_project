```python
# 전체 로우 수 (meta_test: 17267)
import os
import pandas as pd

folder_path = "/Users/parkjisook/Desktop/yeardream/medistream/js/new_json/output/result"

total_row_count = 0  # 전체 행 수를 저장할 변수 초기화

for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)

        try:
            df = pd.read_csv(file_path, low_memory=False)

            row_count = len(df)
            total_row_count += row_count  # 각 파일의 행 수를 더함

        except pd.errors.ParserError as e:
            print(f"파일 '{filename}'에서 ParserError 발생: {e}")

print(f"전체 CSV 파일의 총 로우 수: {total_row_count}")

#17267
```

```python
import pandas as pd
import os

# CSV 파일들이 있는 폴더 경로
folder_path = "/Users/parkjisook/Desktop/yeardream/medistream/js/new_json/output/result"

# 해당 폴더 내의 모든 CSV 파일 가져오기
all_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# 모든 CSV 파일을 담을 빈 DataFrame 생성
combined_df = pd.DataFrame()

# 모든 CSV 파일을 세로로 합치기
for file in all_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    combined_df = pd.concat([combined_df, df], axis=0)

# 결과를 새로운 CSV 파일로 저장
output_path = "/Users/parkjisook/Desktop/yeardream/medistream/js/new_json/output/new_data.csv"
combined_df.to_csv(output_path, index=False)

print(f"CSV 파일이 {output_path}에 성공적으로 저장되었습니다.")
```
